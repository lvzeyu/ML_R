```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
PY100=pd.read_csv("./Data/PY100/PY100.csv")
```

```python
wave1=PY100.loc[:,"sex":"ZQ62"]
```

```python
marriage_status=PY100.loc[:,"IQ45"]
marriage_status
```

```python
wave1["marriage_status"]=marriage_status  
```

```python
wave1["marriage_status"]=wave1["marriage_status"].replace(' ',np.nan)
```

```python
wave1=wave1.dropna(subset=["marriage_status"])
```

```python
imputer = SimpleImputer(strategy='mean')
```

```python
na_values = [8, 9, 88, 99,' ']
for value in na_values:
    wave1.replace(value, np.nan, inplace=True)
```

```python
wave1=wave1.dropna(axis=1, how='all')
```

```python
wave1 = pd.DataFrame(imputer.fit_transform(wave1), columns=wave1.columns)
```

```python
wave1['marriage_status']=wave1['marriage_status'].astype('int')
```

```python
wave1['marriage_status'] = np.where(wave1['marriage_status'] > 1, 1, 0)
```
